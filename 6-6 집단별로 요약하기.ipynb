```python
import pandas as pd
exam = pd.read_csv('exam.csv')
```

```python
exam.agg(mean_math = ('math', 'mean'))
```

```python
#exam['math'].mean()은 전체 평균
#agg()는 전체를 요약한 값을 구하기 보다는 groupby()에 적용해 집단별 요약값을 구할때 사용
```

```python
exam.groupby('nclass').agg(mean_math = ('math', 'mean'))
#반 별 수학 평균
```

```python
#groupby()의 기본값이 변수를 인덱스로 바꾸도록 설정되어 있다.
```

```python
exam.groupby('nclass', as_index = False).agg(mean_math = ('math', 'mean'))
```

```python
exam.groupby('nclass').agg(mean_math = ('math','mean'),
                          sum_math = ('math','sum'),
                          median_math = ('math','median'),
                          n = ('nclass','count')) #빈도 (학생수)
```

```python
mpg = pd.read_csv('mpg.csv')
```

```python
#집단별로 다시 집단 나누기
mpg.groupby(['manufacturer','drv']).agg(mean_cty = ('cty','mean'))
```

```python
#집단별 빈도를 구할 때 groupby()로 데이터를 분리한 다음 agg()에 count()함수를 입력하는 방법을 사용했다.
#하지만, df.value_counts()를 이용하면 집단별 빈도를 더 간단히 구할 수 있다.
#그러나, value_counts()의 경우 출력 결과가 데이터 프레임이 아닌 시리즈 (자료구조)이기 때문에 query()를 적용할 수 없다.
#쿼리를 적용하기 위해선, 데이터 프레임으로 변환하는 과정을 거쳐야 한다.
```

```python
mpg.query('category == "suv"')\
.assign(total = (mpg['hwy']+mpg['cty'])/2)\
.groupby('manufacturer')\
.agg(mean_tot = ('total','mean'))\
.sort_values('mean_tot',ascending = False)\
.head()

#ascending = False : 내림차순 정렬
```
