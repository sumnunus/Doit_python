```python
moon = open('speech_moon.txt', encoding = 'UTF-8').read()
```

```python
moon
```

```python
#불필요한 문자 제거하기
import re #문자처리패키지
moon = re.sub('[^가-힣]', '', moon) #한글이 아닌 모든 문자를 공백으로
#[^가-힣]은 '한글이 아닌 모든 문자'를 의미하는 정규표현식
moon
```

```python
#명사 추출하기
import konlpy
hannanum = konlpy.tag.Hannanum()
nouns = hannanum.nouns(moon)
nouns
```

```python
#데이터 프레임으로 변환
import pandas as pd
df_word = pd.DataFrame({'word' : nouns})
df_word
```

```python
#단어 빈도표 만들기

#글자 수를 나타낸 변수 추가
df_word['count'] = df_word['word'].str.len()
df_word
```

```python
#단어 빈도표 만들기

#두 글자 이상 단어만 남기기
df_word = df_word.query('count >= 2')
df_word.sort_values('count') #오름차순 정렬
```

```python
#단어 빈도 구하기
df_word = df_word.groupby('word', as_index = False)\
.agg(n = ('word','count'))\
.sort_values('n', ascending = False)
df_word
```

```python
#단어 빈도 상위 20개 추출
top20 = df_word.head(20)
top20
```

```python
#그래프 만들기
import seaborn as sns
import matplotlib.pyplot as plt

#한글 폰트, 해상도, 가로 세로 크기 설정
plt.rcParamsm.update({'font.family' : 'Malgun Gothic',
                     'figure.dpi' : '120',
                     'figure.figsize' : [6.5, 6]})

sns.barplot(data = top20, y='word', x='n')
```
